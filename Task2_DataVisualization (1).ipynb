```python
import pandas as pd
import matplotlib.pyplot as plt

# Create sample Superstore-like dataset
data = {
    "Category": ["Technology", "Furniture", "Office Supplies", "Technology", "Furniture", "Office Supplies"],
    "Region": ["West", "East", "South", "West", "East", "South"],
    "Sales": [5000, 3000, 2000, 6000, 3500, 2500],
    "Profit": [800, 400, 200, 900, 450, 250],
    "Order Date": ["2019-01-10", "2020-02-15", "2021-03-20", "2022-04-12", "2023-05-18", "2024-06-25"]
}

df = pd.DataFrame(data)

print("Dataset Loaded Successfully")
print(df)

# Convert date column
df["Order Date"] = pd.to_datetime(df["Order Date"])

# -------------------------
# Visualization 1: Sales by Category
# -------------------------
plt.figure(figsize=(6,4))
df.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Category")
plt.ylabel("Total Sales")
plt.show()

# -------------------------
# Visualization 2: Profit by Region
# -------------------------
plt.figure(figsize=(6,4))
df.groupby("Region")["Profit"].sum().plot(kind="bar")
plt.title("Profit by Region")
plt.ylabel("Profit")
plt.show()

# -------------------------
# Visualization 3: Sales Trend
# -------------------------
sales_trend = df.groupby(df["Order Date"].dt.year)["Sales"].sum()

plt.figure(figsize=(6,4))
sales_trend.plot(marker="o")
plt.title("Yearly Sales Trend")
plt.ylabel("Sales")
plt.show()
```
